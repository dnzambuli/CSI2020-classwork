# The World Bank
## Total GreenHouse Gas Emissions(KT of CO2 equivalent)
Climate Watch Historical GHG Emissions ```1990 - 2022```

### Data selected

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
path = './Total greenhouse gas emissions.csv'
data = pd.read_csv(path)
data.iloc[1:5,]
````

### Data explanation
- The data consists of countries and the total number of drained organic soil carbon dioxide

### clean data


````python
df = data.iloc[3:,].copy()

sect1 = data.iloc[3,:6]
sect2 = data.iloc[3,6:38].astype(int)
sect3 = data.iloc[3,38:]
sect1 = pd.concat([sect1, sect2, sect3])
df.columns = sect1
df.index = range(1, len(df) + 1)
df.iloc[1:3,]
````

### plot data
-- kenya is the country of choice

````python
# select the kenya row
ke = df[df['Country Name'] == 'Kenya']
ke
````

````python
# select the parameter
year = ke.iloc[0,6:36].index
emission = ke.iloc[0,6:36].values
````

````python
# plot
plt.plot(year, emission, marker = 'o')
plt.xlabel('Year')
plt.ylabel('Emission of Co2')
plt.title('Kenya Co2 Emission [KT of Co2]')
plt.show()
````
